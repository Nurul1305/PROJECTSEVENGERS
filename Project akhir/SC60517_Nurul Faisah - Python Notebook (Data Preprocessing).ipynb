```python
# Mengubah Periode Pembayaran menjadi MONTHLY.

def convert_to_monthly_count(row):
    periode_pembayaran = row['periode_pembayaran']
    gaji_minimal = row['gaji_minimal']

    if periode_pembayaran == 'YEARLY':
        return gaji_minimal / 12
    elif periode_pembayaran == 'MONTHLY':
        return gaji_minimal
    elif periode_pembayaran == 'WEEKLY':
        return gaji_minimal * 4
    elif periode_pembayaran == 'HOURLY':
        # Misalkan 40 jam per minggu,
        return gaji_minimal * 40 * 4
    elif periode_pembayaran == 'ONCE':
        return gaji_minimal
    else:
        return gaji_minimal

dfs['gaji_bulanan_minimal'] = dfs.apply(convert_to_monthly_count, axis=1)
```

```python
# Function to create histogram and boxplot.
# This functions takes a dataframe (df) and the variable of interest as arguments.

def diagnostic_plots(dfs, variable):
    # Define figure size
    plt.figure(figsize=(25, 4))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(dfs[variable], bins=30)
    plt.title('Histogram')

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(y=dfs[variable])
    plt.title('Boxplot')

    plt.show()
```

```python
import numpy as np

# Filter numeric columns
numeric_columns = dfs.select_dtypes(include=[np.number]).columns

# Calculate IQR for each numeric column
Q1 = dfs[numeric_columns].quantile(0.25)
Q3 = dfs[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

# Determine upper and lower bounds for each numeric column
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Identify outliers
outliers = ((dfs[numeric_columns] < lower_bound) | (dfs[numeric_columns] > upper_bound)).any(axis=1)

# Display the number of outliers
print("Number of outliers:", outliers.sum())

# Display the rows containing outliers
if outliers.any():
    print("\nRows containing outliers:")
```

```python
# Menghitung IQR untuk setiap kolom numerik
Q1 = dfs[numeric_columns].quantile(0.25)
Q3 = dfs[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

# Tentukan batas atas dan batas bawah untuk setiap kolom numerik
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Identifikasi outlier
outliers = ((dfs[numeric_columns] < lower_bound) | (dfs[numeric_columns] > upper_bound)).any(axis=1)

# Menampilkan jumlah outlier
print("Jumlah outlier:", outliers.sum())

# Menampilkan baris yang mengandung outlier
if outliers.any():
    print("\nBaris yang mengandung outlier:")
    print(dfs[outliers])
else:
    print("\nTidak ada outlier dalam data.")
```

```python
# Let's find outliers in DFf.

diagnostic_plots(dfs, 'gaji_maksimal')
diagnostic_plots(dfs, 'gaji_median')
diagnostic_plots(dfs, 'gaji_minimal')
diagnostic_plots(dfs, 'gaji_bulanan_minimal')
```

```python
# Function to find upper and lower boundaries for skewed distributed variables
# Let's calculate the boundaries outside which sit the outliers for skewed distributions
# Distance passed as an argument, gives us the option to estimate 1.5 times or 3 times the IQR to calculate the boundaries.

def find_skewed_boundaries(dfs, variable, distance):
    IQR = dfs[variable].quantile(0.75) - dfs[variable].quantile(0.25)

    lower_boundary = dfs[variable].quantile(0.25) - (IQR * distance)
    upper_boundary = dfs[variable].quantile(0.75) + (IQR * distance)

    return upper_boundary, lower_boundary
```

```python
# Limits for `gaji_maksimal`
gaji_maksimal_upper_limit, gaji_maksimal_lower_limit = find_skewed_boundaries(dfs, 'gaji_maksimal', 1.5)
print('gaji_maksimal_upper_limit : ', gaji_maksimal_upper_limit)
print('gaji_maksimal_lower_limit : ', gaji_maksimal_lower_limit, '\n')

# Limits for `gaji_median`
gaji_median_upper_limit, gaji_median_lower_limit = find_skewed_boundaries(dfs, 'gaji_median', 1.5)
print('gaji_median_upper_limit : ', gaji_median_upper_limit)
print('gaji_median_lower_limit : ', gaji_median_lower_limit, '\n')

# Limits for `gaji_minimal`
gaji_minimal_upper_limit, gaji_minimal_lower_limit = find_skewed_boundaries(dfs, 'gaji_minimal', 1.5)
print('gaji_minimal_upper_limit : ', gaji_minimal_upper_limit)
print('gaji_minimal_lower_limit : ', gaji_minimal_lower_limit, '\n')

# Limits for `gaji_bulanan_minimal`
gaji_bulanan_minimal_upper_limit, gaji_bulanan_minimal_lower_limit = find_skewed_boundaries(dfs, 'gaji_bulanan_minimal', 1.5)
print('gaji_bulanan_minimal_upper_limit : ', gaji_bulanan_minimal_upper_limit)
print('gaji_bulanan_minimal_lower_limit : ', gaji_bulanan_minimal_lower_limit, '\n')
```

```python
# Flag the outliers in category `gaji_maksimal`
outliers_gaji_maksimal = np.where(dfs['gaji_maksimal'] > gaji_maksimal_upper_limit, True,
                       np.where(dfs['gaji_maksimal'] < gaji_maksimal_lower_limit, True, False))

# Flag the outliers in category `gaji_median`
outliers_gaji_median = np.where(dfs['gaji_median'] > gaji_median_upper_limit, True,
                       np.where(dfs['gaji_median'] < gaji_median_lower_limit, True, False))

# Flag the outliers in category `gaji_minimal`
outliers_gaji_minimal = np.where(dfs['gaji_minimal'] > gaji_minimal_upper_limit, True,
                       np.where(dfs['gaji_minimal'] < gaji_minimal_lower_limit, True, False))

# Flag the outliers in category `gaji_bulanan_minimal`
outliers_gaji_bulanan_minimal = np.where(dfs['gaji_bulanan_minimal'] > gaji_bulanan_minimal_upper_limit, True,
                       np.where(dfs['gaji_bulanan_minimal'] < gaji_bulanan_minimal_lower_limit, True, False))

print(outliers_gaji_maksimal[:10])
```

```python
# Let's trimm the dataset

dfs_trimmed = dfs.loc[~(outliers_gaji_maksimal + outliers_gaji_median + outliers_gaji_minimal +
                            outliers_gaji_bulanan_minimal)]

print('Size dataset - Before trimming : ', dfs.shape)
print('Size dataset - After trimming  : ', dfs_trimmed.shape)
```

```python
print('\ngaji_maksimal - Before Trimming')
diagnostic_plots(dfs, 'gaji_maksimal')
print('\ngaji_maksimal - After Trimming')
diagnostic_plots(dfs_trimmed, 'gaji_maksimal')

print('\ngaji_median - Before Trimming')
diagnostic_plots(dfs, 'gaji_median')
print('\ngaji_median - After Trimming')
diagnostic_plots(dfs_trimmed, 'gaji_median')

print('\ngaji_minimal - Before Trimming')
diagnostic_plots(dfs, 'gaji_minimal')
print('\ngaji_minimal - After Trimming')
diagnostic_plots(dfs_trimmed, 'gaji_minimal')

print('\ngaji_bulanan_minimal - Before Trimming')
diagnostic_plots(dfs, 'gaji_bulanan_minimal')
print('\ngaji_bulanan_minimal - After Trimming')
diagnostic_plots(dfs_trimmed, 'gaji_bulanan_minimal')
```
