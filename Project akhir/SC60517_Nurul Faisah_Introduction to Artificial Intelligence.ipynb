# Lemmatization

Lemmatization untuk mengubah setiap kata dalam teks menjadi bentuk dasarnya. Misalnya, kata "mengalami" akan diubah menjadi "alami", dan kata "berlangsung" akan diubah menjadi "langsung". Dalam beberapa kasus, lemmatization dapat membantu mengurangi dimensi dari representasi teks dengan mengubah kata-kata ke bentuk dasarnya, sehingga memudahkan dalam pemrosesan dan analisis teks.
Alasan mengapa memilih proses lemmatization daripada proses stemming yaitu sebagai berikut:  

1. Lemmatization memiliki hasil yang lebih akurat
  
  Lemmatization mempertimbangkan konteks kata dalam kalimat dan menghasilkan kata dasar yang sesuai dengan makna kata dalam kamus. Sebagai contoh, "better" akan diubah menjadi "good" oleh lemmatization, sedangkan stemming mungkin hanya menghapus akhiran sehingga menghasilkan "bet".
2. Perhatian Terhadap Kelas Kata

  Lemmatization memperhitungkan kelas kata (atau pos-tag) dalam pemrosesan kata, sehingga dapat menghasilkan kata dasar yang sesuai dengan jenis kata tersebut. Misalnya, kata kerja "running" akan diubah menjadi "run", sementara kata benda "better" akan tetap sama. Stemming cenderung hanya menghapus akhiran tanpa memperhatikan jenis kata.
3. Peningkatan Intepretabilitas

  Hasil lemmatization seringkali lebih mudah dipahami daripada hasil stemming, karena kata dasar yang dihasilkan lebih mewakili makna kata sebenarnya.
4. Konsistensi

  Lemmatization menghasilkan kata dasar yang konsisten, sedangkan stemming mungkin menghasilkan kata-kata yang berbeda-beda untuk kata yang sama, tergantung pada aturan stemming yang digunakan.
5. Dukungan Bahasa yang Lebih Baik

  Beberapa bahasa memiliki struktur morfologi yang kompleks, dan lemmatization seringkali lebih efektif dalam menangani variasi kata dalam bahasa-bahasa tersebut.

# Bag of Words

Bag of Words (BoW) adalah representasi model yang digunakan dalam pemrosesan teks, di mana setiap dokumen diwakili sebagai himpunan kata-kata yang ada dalam dokumen tersebut, tanpa memperhatikan struktur tata bahasa atau urutan kata.Kami menggunakan BoW untuk mengonversi setiap teks menjadi representasi vektor yang dapat digunakan dalam analisis lebih lanjut

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Define lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Define the texts
text_medical = "Stunting in children under five refers to the failure of their physical growth, often caused by nutritional deficiencies and infections, especially in the critical growth span from birth to two years of age. According to Asian Development Bank (ADB) data from November 2021, Indonesia is the second highest in Southeast Asia for the prevalence of stunting in children under five, with 31.8%, after Timor Leste. The causes of stunting involve an accumulation of processes throughout life, including pregnancy, childhood and other lifetimes. Factors such as genetic aspects, history of low birth weight, history of infectious diseases, family income level, gender, and nutritional conditions also play a role."
text_sport = "In a fiery match at the Abdullah Bin Khalifa Stadium, the Indonesia U-23 national team made new history in Indonesian football with a 4-1 victory over Jordan, sealing their place in the last 8 of the 2024 U-23 Asian Cup. This victory was not just about numbers, but also about fighting spirit and teamwork. Although names like Marselino Ferdinan, Witan Sulaeman, and Komang Teguh stand out as heroes, this achievement is the result of the hard work and coordination of the team as a whole. The strategy led by coach Shin Tae-yong proved that thorough preparation and the right tactics are the key to success."
text_finance = "Inflation and stagflation are in the media spotlight globally. While they may sound complex, they can have a direct impact on everyday life. Inflation is a general increase in the prices of goods and services over a period of time, while stagflation describes slow economic growth accompanied by continuous price increases. The impacts include a decrease in people's purchasing power and an increase in unemployment. Overcoming this requires good financial management, earning extra income, building up an emergency fund, and making the right investments."

# Tokenize, remove stopwords, and lemmatize for each text
def preprocess_text(text):
    tokens = word_tokenize(text)
    without_stopwords = [word for word in tokens if word not in stop_words]
    lemmatized = [lemmatizer.lemmatize(word) for word in without_stopwords]
    return lemmatized

text_medical_preprocess = preprocess_text(text_medical)
text_sport_preprocess = preprocess_text(text_sport)
text_finance_preprocess = preprocess_text(text_finance)

print(text_medical_preprocess)
print(text_sport_preprocess)
print(text_finance_preprocess)
```

## Menghitung Vektor Bag of Words Pada Topik Teks Kesehatan

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk

# Mendefinisikan text_medical
text_medical = "Stunting in children under five refers to the failure of their physical growth, often caused by nutritional deficiencies and infections, especially in the critical growth span from birth to two years of age. According to Asian Development Bank (ADB) data from November 2021, Indonesia is the second highest in Southeast Asia for the prevalence of stunting in children under five, with 31.8%, after Timor Leste. The causes of stunting involve an accumulation of processes throughout life, including pregnancy, childhood and other lifetimes. Factors such as genetic aspects, history of low birth weight, history of infectious diseases, family income level, gender, and nutritional conditions also play a role."

# Tokenisasi teks
tokens = word_tokenize(text_medical)

# Inisialisasi stop words
stop_words = set(stopwords.words('english'))

# Hapus stop words
without_stopwords = [word for word in tokens if word.lower() not in stop_words]

# Inisialisasi lemmatizer
lemmatizer = WordNetLemmatizer()

# Lematisasi kata-kata
lemmatized = [lemmatizer.lemmatize(word) for word in without_stopwords]

# Gabungkan kembali teks setelah dihilangkan stop words
text = ' '.join(lemmatized)

# Inisialisasi corpus_texts
corpus_texts = [text]

from sklearn.feature_extraction.text import CountVectorizer

# Membuat objek CountVectorizer
cv = CountVectorizer()

# Menghitung vektor Bag of Words
bow_vectors = cv.fit_transform(corpus_texts).toarray()

# Mendapatkan Kata-kata unik (bow_keys)
bow_keys = cv.get_feature_names_out()

# Menampilkan Kata-kata unik
print("\nKata-kata Unik (bow_keys):", bow_keys)

# Menampilkan panjang vektor dan vektor bag of word
print("\nVektor Bag of Words", bow_vectors)
print("\nPanjang Vektor Bag of Word", len(bow_vectors[0]))
```

## Klasifikasi Teks Cristiano Ronaldo

```python
# Mengumpulkan semua teks dalam satu list
texts = [text_medical, text_sport, text_finance]

# Inisialisasi lemmatizer dan stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Membuat list untuk menampung BoW dari setiap teks
corpus_texts = []
for text in texts:
    # Tokenisasi, lemmatisasi, dan penghapusan stop words
    words = word_tokenize(text)
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    text_cleaned = ' '.join(words)
    corpus_texts.append(text_cleaned)

# Membuat BoW dari teks-teks yang telah diproses
cv = CountVectorizer()
bow_vectors = cv.fit_transform(corpus_texts).toarray()

# Normalisasi vektor BoW
bow_texts_norm = []
for bow in bow_vectors:
    length = np.sqrt(np.sum(bow ** 2))
    bow_norm = bow / length
    bow_texts_norm.append(bow_norm)

# Menghitung similarity antara query dan setiap topik
query_text = "Cristiano Ronaldo came off the bench to earn Manchester United a hard-fought 2-1 victory at Everton in the Premier League on Sunday, taking his career goal tally to 700 in the process. Just as United did last weekend in their derby mauling at the hands of local rivals Manchester City, they again found themselves behind early on at Goodison Park after Alex Iwobi curled a sublime strike into the net from 20 metres."
query_words = word_tokenize(query_text)
query_words = [lemmatizer.lemmatize(word) for word in query_words if word not in stop_words]
query_text_cleaned = ' '.join(query_words)
query_bow = cv.transform([query_text_cleaned]).toarray()[0]

similarity_vector = []
for bow in bow_texts_norm:
    similarity = np.dot(bow, query_bow)
    similarity_vector.append(similarity)
print(similarity_vector)

# Mencari topik dengan similarity tertinggi
max_similarity_index = np.argmax(similarity_vector)
if max_similarity_index == 0:
    print("Teks query diklasifikasikan sebagai: Medical")
elif max_similarity_index == 1:
    print("Teks query diklasifikasikan sebagai: Sport")
elif max_similarity_index == 2:
    print("Teks query diklasifikasikan sebagai: Finance")
```
